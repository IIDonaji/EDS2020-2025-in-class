## Selecting a single column
Simple case: selecting a single colum by column name.

General syntax:
```
df[`column_name`]
```

This is an example of **labled-based substting**. which means we select data from our data frame using *name* of the column, *not their* position

## Example

````python
# Select number of bird species observed at Mugu Lagoon in spring
mul_spring = df['MUL_spring']

mul_spring
````

## … using a slice

To select a slice of the columns we will use a special case of `loc` selection 

Sytax:

```
df.loc[ : , 'column_start':'column_end']
```
`column_start` and `column_end` = starting and ending pint of column slice, slice includeds both endpoints.

- the first value to `loc` is used to select rows, using a `:` as the row-slection parameters mean 'select all the rows'

##Example

````python
# Select columns between CSM_winter and Mul_fall
csm_mul = df.loc[:, 'CSM_winter':'MUL_fall']
csm_mul
````

## … using multiple conditions
We can combine multipe conditions to select rows by surrounding each one in parenthesis `()` and using the or operator `|` and the and operator `&`.

````python
# Example: or

# Select rows in CSM with more than 50 species in winter or fall
df[ (df['CSM_winter']>50) | (df['CSM_fall']>50)]
````

Select rows in which both CSM and SDW have more than 60 bord specoes during spring

````python
df[ (df['CSM_spring']>60) & (df['SDW_spring']>60)]
````

### How to select rows and columns simulataneously

Can be done using loc (lables) or iloc (positions)

### ...by lables or conditions

Syntax:
    
```
df.loc[ row-selection, columns-selction]
```

### Example
Select winter surveys for MUGU Lagoon and Tijuana Esturay after 2020

````python
# use `loc`
df.loc[df['year']>2020, ['MUL_winter', 'TJE_winter']]
````

###..by position
Use `iloc`

Syntax:
```
df.iloc row-indices, column-indices]

```
#### Example 
Suppose we want to select rows 3-7 (including 7) and fourth and fifth columns

````python
df.iloc[3:8, [3,4]]
````
